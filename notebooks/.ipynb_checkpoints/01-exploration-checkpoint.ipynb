# 

## 1. Imports



```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import glob
```

# 

## 2. Load dataset

```python
# Use only Friday file for now — small and fast on my PC
df = pd.read_csv(
    "../data/raw/CICIDS2017/Friday-WorkingHours-Afternoon-DDos.pcap_ISCX.csv",
    low_memory=False
)

print(f"Rows: {df.shape[0]:,}")
print(f"Columns: {df.shape[1]}")
```

# 

## 3. Fix known issues

```python
# Fix 1 — remove leading spaces from column names
df.columns = df.columns.str.strip()

# Fix 2 — remove infinite and missing values
df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.dropna(inplace=True)

print("Cleaned!")
print(f"Rows remaining: {df.shape[0]:,}")
```

# 

## 4. Attack distribution

```python
print(df['Label'].value_counts())
```

# 

## 5. Visualisation

```python
df['Label'].value_counts().plot(
    kind='bar',
    figsize=(10, 5),
    color='steelblue',
    edgecolor='black'
)
plt.title('Attack distribution — Friday DDoS file')
plt.xlabel('Attack type')
plt.ylabel('Number of flows')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
